## Training diagnostics

Dump the full training log, save it to `results/`, and plot train vs eval loss and eval macro-F1 per epoch. Used to eyeball overfitting.

```python
import pandas as pd

if MODE == 'quick':
    log = pd.DataFrame(trainer.state.log_history)
    log_path = C.RESULTS_DIR / 'training_log_quick.csv'
    log.to_csv(log_path, index=False)
    print(f'Saved training log: {log_path}')
    display(log.tail(15))
else:
    print('Diagnostics cell below is designed for quick mode. For kfold, per-fold logs live in models/fold_*/')
```

```python
import matplotlib.pyplot as plt

if MODE == 'quick':
    log = pd.read_csv(C.RESULTS_DIR / 'training_log_quick.csv')

    train_rows = log[log['loss'].notna()] if 'loss' in log.columns else log.iloc[0:0]
    eval_rows = log[log['eval_loss'].notna()] if 'eval_loss' in log.columns else log.iloc[0:0]

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # Left: train loss vs eval loss
    if not train_rows.empty:
        axes[0].plot(train_rows['epoch'], train_rows['loss'], marker='o', label='train loss')
    if not eval_rows.empty:
        axes[0].plot(eval_rows['epoch'], eval_rows['eval_loss'], marker='s', label='eval loss')
    axes[0].set_xlabel('epoch'); axes[0].set_ylabel('loss')
    axes[0].set_title('Train vs eval loss')
    axes[0].legend(); axes[0].grid(alpha=0.3)

    # Right: eval macro-F1 per epoch
    if not eval_rows.empty and 'eval_f1_macro' in eval_rows.columns:
        axes[1].plot(eval_rows['epoch'], eval_rows['eval_f1_macro'], marker='o', color='green')
        axes[1].set_xlabel('epoch'); axes[1].set_ylabel('macro-F1')
        axes[1].set_title('Eval macro-F1 per epoch')
        axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig(C.RESULTS_DIR / 'training_curves_quick.png', dpi=150)
    plt.show()

    # Overfitting heuristic: does eval loss turn up while train loss keeps falling?
    if not eval_rows.empty:
        best_eval_epoch = eval_rows.loc[eval_rows['eval_f1_macro'].idxmax(), 'epoch']
        last_epoch = eval_rows['epoch'].max()
        gap = eval_rows['eval_loss'].iloc[-1] - eval_rows['eval_loss'].min()
        print(f'Best eval F1 at epoch {best_eval_epoch:.0f} / {last_epoch:.0f}')
        print(f'Eval loss minimum → final gap: {gap:+.4f}')
        if best_eval_epoch < last_epoch and gap > 0.02:
            print('⚠  Mild overfitting signal: eval loss climbed after the best epoch. Consider fewer epochs.')
        else:
            print('✓  No clear overfitting signal.')
```
